```python
def analyse_engage(likes_list):
    total_likes=0
    for likes in likes_list:
        total_likes=total_likes+likes
    print("Total Likes:",total_likes)
    if total_likes>=1000:
         status= "Viral Post"
    else:
         status= "Normal Engagement"
    print("Post Status:",status)

analyse_engage([300,445])    
```

```python
def analyse_rainfall(rainfall_data,required_level):
    Total_rainfall=0
    for rainfall in rainfall_data:
        Total_rainfall=Total_rainfall+rainfall
    avg=Total_rainfall/len(rainfall_data)
    if avg>=required_level:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"
    print("Average Rainfall:",avg)
    print("Rainfall Status:",status)

analyse_rainfall([78,89,67,65],72)
```

```python
def detect_duplicate(username):
    if len(username)!=len(set(username)):
        result="Yes"
    else:
        result="No"
    print("Duplicate Accounts Found:",result)

detect_duplicate(['user1','user2','user3','user2'])
```

```python
def crop_price(price_list):
    Premium_crops=[]
    for price in price_list:
        if price>2000:
            Premium_crops.append(price)
    print("Premium Crops:",Premium_crops)

crop_price([1800,5000,3500,2000])
```

```python
def health_check(errors):
    if errors==0:
        status="Healthy"
    elif errors<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
        print("Error Count:",errors)
        print("System Status:",status)

health_check(8)
```

```python
def transaction_check(amount):
    limit = 50000
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

```python
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_checker([1,1,0,1,1,1,0,1])
```

```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
