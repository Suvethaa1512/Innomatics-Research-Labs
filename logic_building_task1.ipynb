```python
marks=[45, 78, 90, 33, 60]
pass_count=0
fail_count=0

for i in marks:
    if i>=50:
        pass_count+=1
    else:
        fail_count+=1

print("Total Students Passed:",pass_count)
print("Total Students Failed:",fail_count)
        
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_usernames=[]

for name in names:
    cleaned_name=name.strip().lower()
    cleaned_usernames.append(cleaned_name)

print("Cleaned Usernames:",cleaned_usernames)
```
