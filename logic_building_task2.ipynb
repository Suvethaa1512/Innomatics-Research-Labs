```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee=max(employees,key=employees.get)
print("Highest_Salary:",highest_employee, "-",employees[highest_employee])
```

```python
numbers = [45, 22, 89, 10, 66]
max_value=numbers[0]
min_value=numbers[0]

for num in numbers:
    if num>max_value:
        max_value=num
    if num<min_value:
        min_value=num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
    if price> 1000:
        count+=1
print("Products above 1000:",count)
```

```python
attendance = ["P", "P", "A", "P", "P"]
present_days=attendance.count("P")
total_days=len(attendance)
percentage=(present_days/total_days)*100
print("Attendance Percentage:", percentage)
```
