```python
# Function to calculate average and return result
def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return f"Average: {average} - Pass"
    else:
        return f"Average: {average} - Fail"

# Sample data
student_marks = [60, 45, 70, 55]

print(process_result(student_marks))
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    
    fare = base_fare + (distance * per_km)
    
    if peak == "yes":
        fare += fare * 0.25   # 25% extra
    
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    total_fare = calculate_fare(distance, peak)
    print(f"Total Fare: ₹{total_fare}")
    
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
def check_attendance(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    if percentage >= 75:
        return f"Attendance: {percentage}% - Eligible"
    else:
        return f"Attendance: {percentage}% - Not Eligible"

attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

print(check_attendance(attendance))
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"

password = input("Enter password: ")
print(check_password(password))
```
