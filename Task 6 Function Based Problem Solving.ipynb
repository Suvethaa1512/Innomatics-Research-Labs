```python
def parking(capacity,vehicle_logs):
    parked=0
    peak=0

    for log in vehicle_logs:
        if log=="IN":
            parked+=1
        elif log=="OUT" and parked>0:
            parked-=1
        peak=max(peak,parked)

    status="Available"
    if parked>capacity:
        status="Over capacity"
        
    print("Currently Parked Vehicles:",parked)
    print("Parking Status:",status)

capacity = int(input("Parking Capacity: "))

logs_input = input('Vehicle Logs: ')
vehicle_logs = logs_input.split(",")

parking(capacity, vehicle_logs)
```

```python
def time_estimator(distance,traffic,weather):
    base_time=distance*5

    if traffic=="High":
        base_time+=15
    elif traffic=="Medium":
        base_time+=8

    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")

time_estimator(8,"High","Clear")
```

```python
# Function to analyze seat occupancy
def seat_occupancy(total_seats, booked_seats):

    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)

total_seats = int(input("Total Seats: "))
booked_count = int(input("Number of Booked Seats: "))

booked_seats = [1] * booked_count

seat_occupancy(total_seats, booked_seats)
```

```python
def server_load(readings):
    total=0
    for reading in readings:
        total+=reading
        avg=total/len(readings)

        if avg>80:
            status="Critical"
        elif 50<=avg<=80:
            status="Warning"
        else:
            status="Normal"

    print("Average CPU Load:",round(avg),"%")
    print("Server Status:",status)

reading = eval(input("CPU Readings (%): "))
server_load(reading)

            
```

```python
def classroom_usage(resources):
    overused=[]

    for resource,hours in resources.items():
        if hours>8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_usage(usage)
```

```python
# Function to control registrations
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100, 105)
```
